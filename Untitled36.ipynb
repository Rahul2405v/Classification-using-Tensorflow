```python
def load_decode_jpg(path):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [224, 224])
    image = tf.cast(image, tf.float32) / 255.0
    image.set_shape([224, 224, 3])
    return image
```

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomContrast(0.1),
])
```

```python
def pandas_file(image_name,label,num_classes,path=path):
  image=load_decode_jpg(path+''+image_name)
  label=tf.one_hot(label,num_classes)
  image=augment(image)
  return image,label
```

```python
def augment(image):
    return data_augmentation(image)
```

```python
data=tf.data.Dataset.from_tensor_slices((csvFile['image'], csvFile['labels']))
```

```python
eval_data=tf.data.Dataset.from_tensor_slices((csvFile['image'][1000:1500],csvFile['labels'][1000:1500]))
```

```python
data=(
    data
    .map(lambda x,y: pandas_file(x,y,2,path=path))
    .prefetch(tf.data.AUTOTUNE)
)
```

```python
eval_data=(
    eval_data
    .map(lambda x,y: pandas_file(x,y,2,path=path))
    .prefetch(tf.data.AUTOTUNE)
)
```

```python
model=tf.keras.Sequential(
    [
        tf.keras.layers.Conv2D(filters=32,kernel_size=(3,3),padding='same',activation='relu',input_shape=(224,224,3))
        ,tf.keras.layers.MaxPooling2D(pool_size=(2,2))
        ,tf.keras.layers.Conv2D(filters=64,kernel_size=(3,3),padding='same',activation='relu')
        ,tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(units=128,activation='relu'),
        tf.keras.layers.Dense(units=2,activation='softmax')
    ]
)
```

```python
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy','precision']
)
```

```python
data=(
    data
    .shuffle(1000)
    .batch(32)
    .prefetch(tf.data.AUTOTUNE)
)
```

```python
eval_data=(
    eval_data
    .batch(32)
    .prefetch(tf.data.AUTOTUNE)
)
```

```python
store=model.fit(
    data,
    epochs=5,
    batch_size=32,
    validation_data=eval_data
)
```

```python
import matplotlib.pyplot as plt

plt.plot(store.history['loss'], label='train_loss')
plt.plot(store.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
x = preTrainedModel.output
gap = tf.keras.layers.GlobalAveragePooling2D()(x)
gmp = tf.keras.layers.GlobalMaxPooling2D()(x)
x = tf.keras.layers.Concatenate()([gap, gmp])
x = tf.keras.layers.Dense(128, activation='relu')(x)
output = tf.keras.layers.Dense(2, activation='softmax')(x)
model=Model(inputs=preTrainedModel.input,output=output)
```

```python
model.compile(
    loss='categorical_crossentropy',
    metrics=['accuracy','precision'],
    optimizer='adam'
)
```
